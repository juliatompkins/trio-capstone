```python
import numpy as np
import pandas as pd
```

```python
# Loading data
centroids = pd.read_csv("/Users/dylanmack/Library/CloudStorage/OneDrive-WashingtonUniversityinSt.Louis/ESE 499/trio-capstone/data/CensusTractCentroids.csv")
metro = pd.read_csv("/Users/dylanmack/Library/CloudStorage/OneDrive-WashingtonUniversityinSt.Louis/ESE 499/trio-capstone/data/MetroLinkStations_REGISTERED.csv")
# eventually, get the number of people taking public transportation in each tract to weight the nearest neighbor calculation

centroid_locations = centroids[['INTPTLAT', 'INTPTLON']]
metro_locations = metro[['X', 'Y']]
metro
```

```python
nn_score = 0
closest_metro = np.zeros([len(centroid_locations), 2])
shortest_distance = np.zeros(len(centroid_locations))
for i in range(len(centroid_locations)):
    cent = np.array([centroid_locations.INTPTLAT[i],centroid_locations.INTPTLON[i]])
    closest_distance = 999999
    for j in range(len(metro_locations)):
        metro = np.array([metro_locations.Y[j],metro_locations.X[j]])
        distance = np.linalg.norm(cent-metro)
        if (distance < closest_distance):
            shortest_distance[i] = distance
            closest_distance = distance
            closest_metro[i,0] = metro[0]
            closest_metro[i,1] = metro[1]
            
        
        
    
```
