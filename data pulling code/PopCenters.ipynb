```python
# Function to calculate population-weighted centroid within a polygon
def weighted_centroid(geometry, population):
    mask = geometry_mask([geometry], population_data.shape, transform, invert=True)
    masked_population = np.ma.array(population_data, mask=~mask)
    plot_mask(masked_population)
    total_population = masked_population.sum()
    if total_population > 0:
        weighted_sums_x = np.sum(masked_population.filled(0) * np.arange(masked_population.shape[1]), axis=(0, 1))
        weighted_sums_y = np.sum(masked_population.filled(0) * np.arange(masked_population.shape[0]), axis=(0, 1))
        
        # Calculate centroid coordinates
        centroid_x = np.sum(weighted_sums_x * transform[0]) / total_population + transform[2]
        centroid_y = np.sum(weighted_sums_y * transform[4]) / total_population + transform[5]
        
        return centroid_x, centroid_y
    else:
        centroid = geometry.centroid
        return centroid.x, centroid.y

# Apply the function to calculate population center for each census tract
census_tracts['population_center'] = census_tracts.apply(lambda row: weighted_centroid(row['geometry'], population_data), axis=1)

# Print the census tract ID and its population center
for index, row in census_tracts.iterrows():
    if row['population_center'] is not None:
        print(f'Census Tract {row["NAME"]}: Population Center = {row["population_center"]}')
```
