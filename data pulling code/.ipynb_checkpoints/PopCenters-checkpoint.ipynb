Goal: use population grid and centroids shapefile to get population centers of each tract

```python
#import and read in data 
import geopandas as gpd
import rasterio
from rasterio.features import geometry_mask
import numpy as np
import matplotlib.pyplot as plt
from rasterio.warp import calculate_default_transform, reproject, Resampling
import rioxarray
from rasterio.crs import CRS
from shapely.geometry import Point

#import stuff in as MO projection bc it wasn't playign well with wgs 84 (change back later for compatibility with other scripts)
census_tracts = gpd.read_file('data\census tracts\CensusTractsStl_102696.shp')
population_dataN = rioxarray.open_rasterio('data\StlPop.tif', masked=True).squeeze()

with rasterio.open('data\StlPop.tif') as src:
    transform = src.transform  

print('pop data:',population_dataN.rio.crs)

print('census tracts: ', census_tracts.crs)
population_data = np.nan_to_num(population_dataN, nan=0)
```

```python
## plots to see what we're working with
plt.figure(figsize=(8, 6))
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

im1 = ax1.imshow(population_data, cmap='jet')
ax1.set_title('Population Distribution in STL')
ax1.set_xlabel('Longitude')
ax1.set_ylabel('Latitude')
fig.colorbar(im1, ax=ax1, label='Population')

census_tracts.plot(ax=ax2, edgecolor='red', facecolor='none')
ax2.set_title('Census Tracts in STL')
ax2.set_xlabel('Longitude')
ax2.set_ylabel('Latitude')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
#plot mask function for debugging (is the mask finding anything?)
def plot_mask(mask):
    plt.figure(figsize=(8, 6))
    plt.imshow(mask, cmap='gray', extent=[transform[2], transform[2] + transform[0] * population_data.shape[1], 
                                                 transform[5] + transform[4] * population_data.shape[0], transform[5]])
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title('Mask of First Census Tract')
    plt.grid(False)
    plt.show()
```

```python
# Function to calculate population-weighted centroid within a polygon
def weighted_centroid(geometry, population):
    # mask out each census tract to perform tract-wise coords
    mask = geometry_mask([geometry], population_data.shape, transform, invert=True)
    masked_population = np.ma.array(population_data, mask=~mask)
    # uncomment to observe masked out cenuss tracts 
  #  plot_mask(masked_population)
    total_population = masked_population.sum()
    if total_population > 0:
        # sum population along each axis 
        weighted_sums_x = np.sum(masked_population.filled(0) * np.arange(masked_population.shape[1]), axis=1)
        weighted_sums_y = np.sum(masked_population.filled(0) * np.arange(masked_population.shape[0]).reshape(-1, 1), axis=0)        
        # calc centroid coordinates
        centroid_x = np.sum(weighted_sums_x * transform[0]) / total_population + transform[2]
        centroid_y = np.sum(weighted_sums_y * transform[4]) / total_population + transform[5]
        return centroid_x, centroid_y, False
    else:
        centroid = geometry.centroid
        return centroid.x, centroid.y, True

# save to df 
centroids = []
isZero = []
tract_names = []

for _, row in census_tracts.iterrows():
    center_x, center_y, zero_flag = weighted_centroid(row['geometry'], population_data)
    centroids.append(Point(center_x, center_y))
    isZero.append(zero_flag)
    tract_names.append(row['NAME'])

# the isZero column ensures we have the corrent amount of centroids. The shapely func adds extra fringe ones in (I have no idea why but qgis did this too)
centroids_gdf = gpd.GeoDataFrame(geometry=centroids, crs=census_tracts.crs)
centroids_gdf['isZero'] = isZero  # Add isZero column
centroids_gdf['NAME'] = tract_names

centroids_gdf.to_file('data\PopCenters.shp')
centroids_gdf.head()
```

```python
# removing fringe centroids, includng an extr one in the uppermost tract
centroids_gdf = centroids_gdf[~centroids_gdf['isZero']]
centroids_gdf = centroids_gdf.drop(columns = ['isZero'])
centroids_gdf = centroids_gdf[centroids_gdf.geometry.y <= 1060000]
```

```python
# convert crs back to 4326
final = centroids_gdf.to_crs(epsg='4326')
```

```python
# Plotting the map with centroids to check work
census_tracts_4326 = gpd.read_file('data\census tracts\CensusTractsStl.shp')

plt.figure(figsize=(14, 6))
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# First subplot with population data
im1 = ax1.imshow(population_dataN, cmap='jet')
ax1.set_title('Population Distribution in STL')
ax1.set_xlabel('Longitude')
ax1.set_ylabel('Latitude')
fig.colorbar(im1, ax=ax1, label='Population')

# Second subplot with census tracts and centroids
census_tracts_4326.plot(ax=ax2, edgecolor='red', facecolor='none')
ax2.set_title('Census Tracts in STL')
ax2.set_xlabel('Longitude')
ax2.set_ylabel('Latitude')

# Plot centroids based on isZero column
final.plot(ax=ax2, color='green', markersize=10, label = 'Population Centers')

# Display the legend for centroids
ax2.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
